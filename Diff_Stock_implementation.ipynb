```python
# Calculate turbulence index
def calculate_turbulence_index(df, window=252):
    turbulence_index = pd.DataFrame(index=df.index)
    for ticker in df.columns.levels[0]:
        returns = df[(ticker, 'Close')].pct_change().dropna()
        rolling_mean = returns.rolling(window=window).mean()
        rolling_cov = returns.rolling(window=window).cov()
        turbulence = []
        for i in range(window, len(returns)):
            mean_diff = returns.iloc[i] - rolling_mean.iloc[i]
            cov_matrix = rolling_cov.iloc[i - window + 1:i + 1]
            if cov_matrix.shape[0] == window:
                cov_matrix = cov_matrix.values.reshape(window, -1)
                try:
                    inv_cov_matrix = np.linalg.inv(np.cov(cov_matrix, rowvar=False))
                    turbulence_value = mean_diff.T @ inv_cov_matrix @ mean_diff
                    turbulence.append(turbulence_value)
                except np.linalg.LinAlgError:
                    turbulence.append(np.nan)
        turbulence_index[ticker] = pd.Series(turbulence, index=returns.index[window:])
    return turbulence_index

def calculate_turbulence_threshold(turbulence_index, percentile=90):
    threshold = np.percentile(turbulence_index.values.flatten(), percentile)
    return threshold
```

```python
class StockTradingEnv(gym.Env):
    def __init__(self, df, embeddings, turbulence_index, window_size=10, turbulence_threshold=None, h_max=100):
        super(StockTradingEnv, self).__init__()
        self.df = df
        self.embeddings = embeddings
        self.turbulence_index = turbulence_index
        self.window_size = window_size
        self.turbulence_threshold = turbulence_threshold
        self.h_max = h_max

        self.min_reward = float('inf')
        self.max_reward = float('-inf')
        self.total_value = 1000000  # Starting with 1M balance
        self.daily_returns = []
        self.risk_free_rate = 0.01 / 252  # Annual risk-free rate, converted to daily

        self.action_space = spaces.Box(
            low=-1, high=1, shape=(30,), dtype=np.float32
        )

        self.df = self.add_indicators(self.df)
        self.num_features_per_stock = len(['Adj Close', 'MACD', 'MACD_Signal', 'MACD_Hist', 'RSI', 'CCI', 'ADX'])
        embedding_size = len(self.embeddings.iloc[0]['embedding'])
        self.observation_space = spaces.Box(
            low=-np.inf,
            high=np.inf,
            shape=(self.window_size * self.num_features_per_stock * 30 + self.window_size * embedding_size,),
            dtype=np.float64
        )

        self.reset()

    def add_indicators(self, df):
        for ticker in df.columns.levels[0]:
            df[(ticker, 'MACD')], df[(ticker, 'MACD_Signal')], df[(ticker, 'MACD_Hist')] = talib.MACD(df[(ticker, 'Close')], fastperiod=12, slowperiod=26, signalperiod=9)
            df[(ticker, 'RSI')] = talib.RSI(df[(ticker, 'Close')], timeperiod=14)
            df[(ticker, 'CCI')] = talib.CCI(df[(ticker, 'High')], df[(ticker, 'Low')], df[(ticker, 'Close')], timeperiod=14)
            df[(ticker, 'ADX')] = talib.ADX(df[(ticker, 'High')], df[(ticker, 'Low')], df[(ticker, 'Close')], timeperiod=14)
        return df

    def reset(self, seed=None, options=None):
        self.current_step = 0
        self.balance = 1000000
        self.shares_held = {ticker: 0 for ticker in self.df.columns.levels[0]}
        self.total_value = self.balance
        self.actions = []  # Store actions taken
        self.daily_returns = []  # Reset daily returns
        self.total_values = []  # Store values
        return self._next_observation(), {}

    def _next_observation(self):
        start = max(0, self.current_step - self.window_size + 1)
        end = self.current_step + 1

        num_stocks = len(self.df.columns.levels[0])
        obs = np.empty((self.window_size, self.num_features_per_stock * num_stocks + len(self.embeddings.iloc[0]['embedding'])))
        for i, ticker in enumerate(self.df.columns.levels[0]):
            ticker_data = self.df[ticker].iloc[start:end]
            if len(ticker_data) < self.window_size:
                ticker_data = ticker_data.reindex(pd.date_range(start=ticker_data.index[0], periods=self.window_size, freq='D'), fill_value=0)

            adj_close = ticker_data['Adj Close'].values
            macd, macd_signal, macd_hist = talib.MACD(ticker_data['Close'].values, fastperiod=12, slowperiod=26, signalperiod=9)
            rsi = talib.RSI(ticker_data['Close'].values, timeperiod=14)
            cci = talib.CCI(ticker_data['High'].values, ticker_data['Low'].values, ticker_data['Close'].values, timeperiod=14)
            adx = talib.ADX(ticker_data['High'].values, ticker_data['Low'].values, ticker_data['Close'].values, timeperiod=14)

            adj_close = np.nan_to_num(adj_close)
            macd = np.nan_to_num(macd)
            macd_signal = np.nan_to_num(macd_signal)
            macd_hist = np.nan_to_num(macd_hist)
            rsi = np.nan_to_num(rsi)
            cci = np.nan_to_num(cci)
            adx = np.nan_to_num(adx)

            features = np.column_stack([adj_close, macd, macd_signal, macd_hist, rsi, cci, adx])
            obs[:, i * self.num_features_per_stock: (i + 1) * self.num_features_per_stock] = features

        for j in range(self.window_size):
            if start + j < len(self.df.index):
                date = self.df.index[start + j]
                news_embedding = self.embeddings.loc[date]['embedding']
                obs[j, self.num_features_per_stock * num_stocks:] = news_embedding
        return obs.flatten()

    def step(self, action):
        self.current_step += 1
        self.actions.append(action)  # Record the action

        action = (action * self.h_max).astype(int)

        terminated = self.current_step >= len(self.df) - 1
        truncated = False
        prev_total_value = self.total_value

        for i, ticker in enumerate(self.df.columns.levels[0]):
          close_price = self.df[ticker]['Close'].iloc[self.current_step]
          if close_price != 0:
              if action[i] > 0:
                  shares_to_buy = min(self.balance // close_price, action[i])
                  self.shares_held[ticker] += shares_to_buy
                  self.balance -= shares_to_buy * close_price
              elif action[i] < 0:
                  shares_to_sell = min(abs(action[i]), self.shares_held[ticker])
                  self.shares_held[ticker] -= shares_to_sell
                  self.balance += shares_to_sell * close_price

        self.total_value = self.balance + sum(self.shares_held[ticker] * self.df[ticker]['Close'].iloc[self.current_step] for ticker in self.df.columns.levels[0])
        transaction_cost = 0.001 * sum(self.df[ticker]['Close'].iloc[self.current_step] * abs(action[i]) for i, ticker in enumerate(self.df.columns.levels[0]))
        reward = (self.total_value - prev_total_value) - transaction_cost

        # Calculate daily return and store it
        daily_return = (self.total_value - prev_total_value) / prev_total_value
        self.daily_returns.append(daily_return)

        # Calculate maximum drawdown
        drawdown_penalty = self._calculate_drawdown_penalty()

        reward -= drawdown_penalty

        self.total_value -= transaction_cost
        self.total_values.append(self.total_value)

        self.min_reward = min(self.min_reward, reward)
        self.max_reward = max(self.max_reward, reward)
        normalized_reward = self._normalize_reward(reward)

        return self._next_observation(), normalized_reward, terminated, truncated, {"actual_reward": reward}

    def _calculate_drawdown_penalty(self):
        peak = max(self.total_values, default=self.total_value)
        drawdown = (peak - self.total_value) / peak if peak != 0 else 0
        drawdown_penalty = 0.01 * drawdown
        return drawdown_penalty

    def _normalize_reward(self, reward):
        if self.max_reward == self.min_reward:
            return 0
        return (reward - self.min_reward) / (self.max_reward - self.min_reward)

    def compute_final_reward(self):
        sharpe_ratio = calculate_sharpe_ratio(self.daily_returns, self.risk_free_rate)
        final_reward = self.total_value - 1000000  # Profit relative to initial investment
        final_reward += sharpe_ratio * 1000  # Adjust the multiplier as needed to balance the reward components
        return final_reward

    def render(self, mode='human'):
        profit = self.total_value - 1000000
        print(f'Step: {self.current_step}, Balance: {self.balance}, Total Value: {self.total_value}, Profit: {profit}')

    def get_actions(self):
        return self.actions

    def get_total_values(self):
        return self.total_values
```

```python
def calculate_sharpe_ratio(daily_returns, risk_free_rate=0.0):
    mean_return = np.mean(daily_returns)
    std_return = np.std(daily_returns)
    if std_return == 0:
        return 0
    sharpe_ratio = (mean_return - risk_free_rate) / std_return * np.sqrt(252)  # Annualize the ratio
    return sharpe_ratio
```

```python
import os
import numpy as np
from collections import deque
import random

class RandomInitialReplayMemory:
    def __init__(self, capacity=100000):
        self.buffer = deque(maxlen=capacity)

    def push(self, state, action, reward, next_state, done):
        self.buffer.append((state, action, reward, next_state, done))

    def sample(self, batch_size):
        batch = random.sample(self.buffer, batch_size)
        states, actions, rewards, next_states, dones = map(np.array, zip(*batch))
        return states, actions, rewards, next_states, dones

    def __len__(self):
        return len(self.buffer)

    def save(self, filename):
        """Save replay memory as npz file."""
        states, actions, rewards, next_states, dones = zip(*self.buffer)
        np.savez_compressed(
            filename,
            states=np.array(states, dtype=np.float32),
            actions=np.array(actions, dtype=np.float32),
            rewards=np.array(rewards, dtype=np.float32),
            next_states=np.array(next_states, dtype=np.float32),
            dones=np.array(dones, dtype=np.bool_)
        )
        print(f"Replay memory saved to {filename}")

    def load(self, filename):
        """Load replay memory from npz file."""
        data = np.load(filename)
        self.buffer = deque(
            zip(
                data["states"],
                data["actions"],
                data["rewards"],
                data["next_states"],
                data["dones"],
            ),
            maxlen=len(data["states"])
        )
        print(f"Replay memory loaded from {filename}")


def build_random_replay_memory(env, steps=10000, capacity=100000, save_path="drive/MyDrive/replay_memory.npz"):
    memory = RandomInitialReplayMemory(capacity=capacity)
    state, _ = env.reset()

    for step in range(steps):
        action = env.action_space.sample()
        next_state, reward, done, truncated, info = env.step(action)
        memory.push(state, action, reward, next_state, done)
        state = next_state

        if done:
            state, _ = env.reset()

        if (step + 1) % 1000 == 0:
            print(f"Collected {step + 1}/{steps} transitions.")

    # Save replay memory
    os.makedirs(os.path.dirname(save_path), exist_ok=True)
    memory.save(save_path)
    return memory


memory = build_random_replay_memory(env, steps=10000, save_path="drive/MyDrive/stock_project/RM1.npz")

# # Later:
# memory = ReplayMemory()
# memory.load("drive/MyDrive/stock_project/RM1.npz")
```

```python
import os
import numpy as np
from collections import deque
import random

class ReplayMemory:
    """
    Prioritized Experience Replay (PER) based on TD-error.
    API-compatible with your usage: push/save/load and sample.
    - sample(...) returns (states, actions, next_states, rewards, not_dones, weights, idxs)
      so we can weight the critic loss and update priorities.
    """
    def __init__(self, capacity=100000, alpha=0.6, beta=0.4, beta_increment_per_sampling=1e-3, eps=1e-6):
        self.capacity = capacity
        self.alpha = alpha
        self.beta = beta
        self.beta_inc = beta_increment_per_sampling
        self.eps = eps

        self.ptr = 0
        self.full = False

        self.states = None
        self.actions = None
        self.rewards = None
        self.next_states = None
        self.dones = None
        self.priorities = np.zeros((capacity,), dtype=np.float32)

    def _init_arrays(self, state, action, reward, next_state, done):
        S = state.shape[0]; A = action.shape[0]
        self.states      = np.zeros((self.capacity, S), dtype=np.float32)
        self.actions     = np.zeros((self.capacity, A), dtype=np.float32)
        self.rewards     = np.zeros((self.capacity, 1), dtype=np.float32)
        self.next_states = np.zeros((self.capacity, S), dtype=np.float32)
        self.dones       = np.zeros((self.capacity, 1), dtype=np.bool_)

    def push(self, state, action, reward, next_state, done, priority=None):
        # ensure numpy dtypes
        state = np.asarray(state, dtype=np.float32)
        action = np.asarray(action, dtype=np.float32)
        reward = np.asarray(reward, dtype=np.float32).reshape(1,)
        next_state = np.asarray(next_state, dtype=np.float32)
        done = np.asarray(done, dtype=np.bool_)

        if self.states is None:
            self._init_arrays(state, action, reward, next_state, done)

        i = self.ptr
        self.states[i]      = state
        self.actions[i]     = action
        self.rewards[i]     = reward
        self.next_states[i] = next_state
        self.dones[i]       = done

        if priority is None:
            p = self.priorities.max() if (i > 0 or self.full) else 1.0
        else:
            p = float(priority)
        self.priorities[i] = p

        self.ptr = (self.ptr + 1) % self.capacity
        if self.ptr == 0:
            self.full = True

    def sample(self, batch_size):
        N = self.capacity if self.full else self.ptr
        probs = (self.priorities[:N] + self.eps) ** self.alpha
        probs = probs / probs.sum()

        replace_flag = False if N >= batch_size else True
        idxs = np.random.choice(N, size=batch_size, p=probs, replace=replace_flag)

        # Importance Sampling weights
        self.beta = min(1.0, self.beta + self.beta_inc)
        weights = (N * probs[idxs]) ** (-self.beta)
        weights = weights / weights.max()
        weights = weights.astype(np.float32).reshape(-1, 1)

        states      = torch.tensor(self.states[idxs], dtype=torch.float32)
        actions     = torch.tensor(self.actions[idxs], dtype=torch.float32)
        rewards     = torch.tensor(self.rewards[idxs], dtype=torch.float32)
        next_states = torch.tensor(self.next_states[idxs], dtype=torch.float32)
        not_dones   = torch.tensor(1.0 - self.dones[idxs].astype(np.float32), dtype=torch.float32)
        weights_t   = torch.tensor(weights, dtype=torch.float32)
        idxs_t      = torch.tensor(idxs, dtype=torch.long)
        return states, actions, next_states, rewards, not_dones, weights_t, idxs_t

    def update_priorities(self, idxs, td_errors):
        te = td_errors.detach().abs().cpu().numpy().reshape(-1) + self.eps
        self.priorities[idxs.cpu().numpy()] = te

    def __len__(self):
        return self.capacity if self.full else self.ptr

    def save(self, filename):
        N = self.capacity if self.full else self.ptr
        np.savez_compressed(
            filename,
            states=self.states[:N],
            actions=self.actions[:N],
            rewards=self.rewards[:N],
            next_states=self.next_states[:N],
            dones=self.dones[:N],
            priorities=self.priorities[:N]
        )
        print(f"Replay memory (PER) saved to {filename} ({N} transitions)")

    def load(self, filename):
        data = np.load(filename)
        N = data["states"].shape[0]
        S = data["states"].shape[1]
        A = data["actions"].shape[1]
        self.capacity = max(self.capacity, N)

        # re-init with capacity
        self.states      = np.zeros((self.capacity, S), dtype=np.float32)
        self.actions     = np.zeros((self.capacity, A), dtype=np.float32)
        self.rewards     = np.zeros((self.capacity, 1), dtype=np.float32)
        self.next_states = np.zeros((self.capacity, S), dtype=np.float32)
        self.dones       = np.zeros((self.capacity, 1), dtype=np.bool_)
        self.priorities  = np.zeros((self.capacity,), dtype=np.float32)

        self.states[:N]      = data["states"]
        self.actions[:N]     = data["actions"]
        self.rewards[:N]     = data["rewards"]
        self.next_states[:N] = data["next_states"]
        self.dones[:N]       = data["dones"]
        if "priorities" in data:
            self.priorities[:N] = data["priorities"]
        else:
            self.priorities[:N] = 1.0  # uniform if old file
        self.ptr = N % self.capacity
        self.full = (N == self.capacity)
        print(f"Replay memory (PER) loaded from {filename} ({N} transitions)")


```

```python
# ==== shape_helpers.py ====
import torch

class StateShaper:
    """
    Reshapes flat state vectors into [B, T, F] for sequence encoders.
    Assumes a fixed window_size T (default 10, matching your env).
    """
    def __init__(self, window_size: int = 10, per_step_dim: int = None, state_dim: int = None):
        self.T = window_size
        if per_step_dim is None and state_dim is not None:
            assert state_dim % self.T == 0, f"state_dim {state_dim} not divisible by T={self.T}"
            per_step_dim = state_dim // self.T
        self.F = per_step_dim

    def to_sequence(self, state_flat: torch.Tensor) -> torch.Tensor:
        # state_flat: [B, T*F]  -> [B, T, F]
        B = state_flat.shape[0]
        if self.F is None:
            # infer F on first call
            total = state_flat.shape[1]
            assert total % self.T == 0, f"state length {total} not divisible by T={self.T}"
            self.F = total // self.T
        return state_flat.view(B, self.T, self.F)
```

```python
# ==== helpers.py (minimal, no external deps) ====
import math
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F

class SinusoidalPosEmb(nn.Module):
    def __init__(self, dim: int):
        super().__init__()
        self.dim = dim

    def forward(self, x):
        device = x.device
        half_dim = self.dim // 2
        emb = math.log(10000) / (half_dim - 1)
        emb = torch.exp(torch.arange(half_dim, device=device) * -emb)
        emb = x[:, None] * emb[None, :]
        return torch.cat((emb.sin(), emb.cos()), dim=-1)

def extract(a, t, x_shape):
    # a: [T], t: [B], return broadcast to x_shape
    b, *_ = t.shape
    out = a.gather(-1, t)
    return out.reshape(b, *((1,) * (len(x_shape) - 1)))

def cosine_beta_schedule(timesteps, s=0.008, dtype=torch.float32):
    steps = timesteps + 1
    x = np.linspace(0, steps, steps)
    alphas_cumprod = np.cos(((x / steps) + s) / (1 + s) * np.pi * 0.5) ** 2
    alphas_cumprod = alphas_cumprod / alphas_cumprod[0]
    betas = 1 - (alphas_cumprod[1:] / alphas_cumprod[:-1])
    betas_clipped = np.clip(betas, a_min=0, a_max=0.999)
    return torch.tensor(betas_clipped, dtype=dtype)

def linear_beta_schedule(timesteps, beta_start=1e-4, beta_end=2e-2, dtype=torch.float32):
    betas = np.linspace(beta_start, beta_end, timesteps)
    return torch.tensor(betas, dtype=dtype)

def vp_beta_schedule(timesteps, dtype=torch.float32):
    t = np.arange(1, timesteps + 1)
    T = timesteps
    b_max, b_min = 10., 0.1
    alpha = np.exp(-b_min / T - 0.5 * (b_max - b_min) * (2 * t - 1) / T ** 2)
    betas = 1 - alpha
    return torch.tensor(betas, dtype=dtype)

# Weighted losses
class WeightedLoss(nn.Module):
    def forward(self, pred, targ, weights=1.0):
        loss = self._loss(pred, targ)
        return (loss * weights).mean()

class WeightedL1(WeightedLoss):
    def _loss(self, pred, targ): return torch.abs(pred - targ)

class WeightedL2(WeightedLoss):
    def _loss(self, pred, targ): return F.mse_loss(pred, targ, reduction='none')

Losses = {'l1': WeightedL1, 'l2': WeightedL2}
```

```python
# ==== lstm_state_encoder.py ====
import torch
import torch.nn as nn

class LSTMStateEncoder(nn.Module):
    """
    Encodes [B,T,F] -> [B,H] using (Bi)LSTM + (last+mean)/2 pooling.
    """
    def __init__(self, input_dim: int, hidden: int = 256, layers: int = 1, bidirectional: bool = True, dropout: float = 0.1):
        super().__init__()
        self.lstm = nn.LSTM(
            input_size=input_dim,
            hidden_size=hidden,
            num_layers=layers,
            batch_first=True,
            bidirectional=bidirectional,
            dropout=dropout if layers > 1 else 0.0
        )
        self.out_dim = hidden * (2 if bidirectional else 1)
        self.norm = nn.LayerNorm(self.out_dim)
        self.dropout = nn.Dropout(dropout)

    def forward(self, seq: torch.Tensor) -> torch.Tensor:
        out, _ = self.lstm(seq)           # [B,T,H*(1 or 2)]
        last = out[:, -1, :]
        mean = out.mean(dim=1)
        h = 0.5 * (last + mean)
        return self.dropout(self.norm(h)) # [B,out_dim]
```

```python
# ==== model.py (policy backbone) ====
import torch
import torch.nn as nn
import torch.nn.functional as F

class MLP(nn.Module):
    """
    Same class name & forward signature as before:
    f(x_noisy, time, state_flat) -> predicted noise.
    Internally: state_flat â†’ [B,T,F] â†’ (Bi)LSTM â†’ [B,H] â†’ concat with x_noisy & t_emb.
    """
    def __init__(self, state_dim: int, action_dim: int, device: str, t_dim: int = 16,
                 window_size: int = 10, bidirectional: bool = True):
        super().__init__()
        self.device = device
        self.window_size = window_size
        self.bidirectional = bidirectional

        # time embedding (unchanged)
        self.time_mlp = nn.Sequential(
            SinusoidalPosEmb(t_dim),
            nn.Linear(t_dim, t_dim * 2),
            nn.Mish(),
            nn.Linear(t_dim * 2, t_dim),
        )

        # sequence shaping + encoder
        self.shaper = StateShaper(window_size=window_size, state_dim=state_dim)
        per_step_dim = self.shaper.F
        self.state_encoder = LSTMStateEncoder(
            input_dim=per_step_dim, hidden=256, layers=1, bidirectional=bidirectional, dropout=0.1
        )
        enc_dim = self.state_encoder.out_dim

        # MLP head now consumes: [x_noisy, t_emb, encoded_state]
        input_dim = action_dim + t_dim + enc_dim
        self.mid_layer = nn.Sequential(
            nn.Linear(input_dim, 256), nn.Mish(),
            nn.Linear(256, 256), nn.Mish(),
            nn.Linear(256, 256), nn.Mish(),
        )
        self.final_layer = nn.Linear(256, action_dim)

    def forward(self, x, time, state):
        # x: [B, action_dim]; time: [B] long/int; state: [B, state_dim] (flat)
        t = self.time_mlp(time)
        seq = self.shaper.to_sequence(state)          # [B,T,F]
        state_enc = self.state_encoder(seq)           # [B,H]
        h = torch.cat([x, t, state_enc], dim=1)
        h = self.mid_layer(h)
        return self.final_layer(h)
```

```python
# ==== diffusion.py (core) ====
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F

# from helpers cell
#   cosine_beta_schedule, linear_beta_schedule, vp_beta_schedule, extract, Losses

class Diffusion(nn.Module):
    def __init__(self, state_dim, action_dim, model, max_action,
                 beta_schedule='linear', n_timesteps=100,
                 loss_type='l2', clip_denoised=True, predict_epsilon=True):
        super().__init__()
        self.state_dim = state_dim
        self.action_dim = action_dim
        self.max_action = max_action
        self.model = model

        if beta_schedule == 'linear':
            betas = linear_beta_schedule(n_timesteps)
        elif beta_schedule == 'cosine':
            betas = cosine_beta_schedule(n_timesteps)
        elif beta_schedule == 'vp':
            betas = vp_beta_schedule(n_timesteps)
        else:
            raise ValueError(f"Unknown beta_schedule: {beta_schedule}")

        alphas = 1. - betas
        alphas_cumprod = torch.cumprod(alphas, axis=0)
        alphas_cumprod_prev = torch.cat([torch.ones(1), alphas_cumprod[:-1]])

        self.n_timesteps = int(n_timesteps)
        self.clip_denoised = clip_denoised
        self.predict_epsilon = predict_epsilon

        # Register buffers
        self.register_buffer('betas', betas)
        self.register_buffer('alphas_cumprod', alphas_cumprod)
        self.register_buffer('alphas_cumprod_prev', alphas_cumprod_prev)
        self.register_buffer('sqrt_alphas_cumprod', torch.sqrt(alphas_cumprod))
        self.register_buffer('sqrt_one_minus_alphas_cumprod', torch.sqrt(1. - alphas_cumprod))
        self.register_buffer('log_one_minus_alphas_cumprod', torch.log(1. - alphas_cumprod))
        self.register_buffer('sqrt_recip_alphas_cumprod', torch.sqrt(1. / alphas_cumprod))
        self.register_buffer('sqrt_recipm1_alphas_cumprod', torch.sqrt(1. / alphas_cumprod - 1))

        posterior_variance = betas * (1. - alphas_cumprod_prev) / (1. - alphas_cumprod)
        self.register_buffer('posterior_variance', posterior_variance)
        self.register_buffer('posterior_log_variance_clipped', torch.log(torch.clamp(posterior_variance, min=1e-20)))
        self.register_buffer('posterior_mean_coef1', betas * torch.sqrt(alphas_cumprod_prev) / (1. - alphas_cumprod))
        self.register_buffer('posterior_mean_coef2', (1. - alphas_cumprod_prev) * torch.sqrt(alphas) / (1. - alphas_cumprod))

        self.loss_fn = Losses[loss_type]()

    # ----- sampling -----
    def predict_start_from_noise(self, x_t, t, noise):
        if self.predict_epsilon:
            return (
                extract(self.sqrt_recip_alphas_cumprod, t, x_t.shape) * x_t -
                extract(self.sqrt_recipm1_alphas_cumprod, t, x_t.shape) * noise
            )
        else:
            return noise

    def q_posterior(self, x_start, x_t, t):
        posterior_mean = (
            extract(self.posterior_mean_coef1, t, x_t.shape) * x_start +
            extract(self.posterior_mean_coef2, t, x_t.shape) * x_t
        )
        posterior_variance = extract(self.posterior_variance, t, x_t.shape)
        posterior_log_variance_clipped = extract(self.posterior_log_variance_clipped, t, x_t.shape)
        return posterior_mean, posterior_variance, posterior_log_variance_clipped

    def p_mean_variance(self, x, t, state):
        x_recon = self.predict_start_from_noise(x, t=t, noise=self.model(x, t, state))
        if self.clip_denoised:
            x_recon.clamp_(-self.max_action, self.max_action)
        model_mean, posterior_variance, posterior_log_variance = self.q_posterior(
            x_start=x_recon, x_t=x, t=t)
        return model_mean, posterior_variance, posterior_log_variance

    def p_sample(self, x, t, state):
        b, *_, device = *x.shape, x.device
        model_mean, _, model_log_variance = self.p_mean_variance(x=x, t=t, state=state)
        noise = torch.randn_like(x)
        nonzero_mask = (1 - (t == 0).float()).reshape(b, *((1,) * (len(x.shape) - 1)))
        return model_mean + nonzero_mask * (0.5 * model_log_variance).exp() * noise

    @torch.no_grad()
    def p_sample_loop(self, state, shape):
        device = self.betas.device
        batch_size = shape[0]
        x = torch.randn(shape, device=device)
        for i in reversed(range(0, self.n_timesteps)):
            t = torch.full((batch_size,), i, device=device, dtype=torch.long)
            x = self.p_sample(x, t, state)
        return x

    @torch.no_grad()
    def sample(self, state):
        B = state.shape[0]
        action = self.p_sample_loop(state, (B, self.action_dim))
        return action.clamp_(-self.max_action, self.max_action)

    # ----- training -----
    def q_sample(self, x_start, t, noise=None):
        if noise is None:
            noise = torch.randn_like(x_start)
        return (
            extract(self.sqrt_alphas_cumprod, t, x_start.shape) * x_start +
            extract(self.sqrt_one_minus_alphas_cumprod, t, x_start.shape) * noise
        )

    def p_losses(self, x_start, state, t, weights=1.0):
        noise = torch.randn_like(x_start)
        x_noisy = self.q_sample(x_start=x_start, t=t, noise=noise)
        x_recon = self.model(x_noisy, t, state)
        if self.predict_epsilon:
            loss = self.loss_fn(x_recon, noise, weights)
        else:
            loss = self.loss_fn(x_recon, x_start, weights)
        return loss

    def loss(self, x, state, weights=1.0):
        B = len(x)
        t = torch.randint(0, self.n_timesteps, (B,), device=x.device).long()
        return self.p_losses(x, state, t, weights)

    def forward(self, state):
        return self.sample(state)
```

```python
class EMA:
    def __init__(self, beta=0.995):
        self.beta = beta
    def update_model_average(self, ma_model, current_model):
        for p_cur, p_ma in zip(current_model.parameters(), ma_model.parameters()):
            p_ma.data.mul_(self.beta).add_(p_cur.data, alpha=(1.0 - self.beta))
```

```python
# ==== ql_diffusion.py (standalone agent) ====
import copy
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F

# from previous cells
#   MLP, Diffusion

class Critic(nn.Module):
    """
    Same class name & methods.
    Internally encodes state_flat via (Bi)LSTM over [B,T,F] and then concatenates with action.
    """
    def __init__(self, state_dim, action_dim, hidden_dim=256, window_size: int = 10, bidirectional: bool = True):
        super().__init__()
        self.shaper = StateShaper(window_size=window_size, state_dim=state_dim)
        per_step_dim = self.shaper.F
        self.encoder = LSTMStateEncoder(
            input_dim=per_step_dim, hidden=256, layers=1, bidirectional=bidirectional, dropout=0.1
        )
        enc_dim = self.encoder.out_dim

        def q_net():
            return nn.Sequential(
                nn.Linear(enc_dim + action_dim, hidden_dim), nn.Mish(),
                nn.Linear(hidden_dim, hidden_dim), nn.Mish(),
                nn.Linear(hidden_dim, hidden_dim), nn.Mish(),
                nn.Linear(hidden_dim, 1),
            )
        self.q1_model = q_net()
        self.q2_model = q_net()

    def forward(self, state, action):
        seq = self.shaper.to_sequence(state)      # [B,T,F]
        z = self.encoder(seq)                     # [B,enc_dim]
        x = torch.cat([z, action], dim=-1)
        return self.q1_model(x), self.q2_model(x)

    def q_min(self, state, action):
        q1, q2 = self.forward(state, action)
        return torch.min(q1, q2)


class Diffusion_QL(object):
    def __init__(self,
                 state_dim,
                 action_dim,
                 max_action,
                 device,
                 discount=0.99,
                 tau=0.005,
                 max_q_backup=False,
                 eta=0.5,
                 beta_schedule='cosine',
                 n_timesteps=1000,
                 ema_decay=0.995,
                 step_start_ema=1000,
                 update_ema_every=5,
                 lr=3e-4,
                 lr_decay=False,
                 lr_maxt=1000,
                 grad_norm=1.0):
        # Policy (diffusion)
        self.model = MLP(state_dim=state_dim, action_dim=action_dim, device=device)
        self.actor = Diffusion(state_dim=state_dim, action_dim=action_dim,
                               model=self.model, max_action=max_action,
                               beta_schedule=beta_schedule, n_timesteps=n_timesteps).to(device)
        self.actor_optimizer = torch.optim.Adam(self.actor.parameters(), lr=lr)

        # EMA policy for target actions
        self.step = 0
        self.step_start_ema = step_start_ema
        self.ema = EMA(ema_decay)
        self.ema_model = copy.deepcopy(self.actor)
        self.update_ema_every = update_ema_every

        # Critic
        self.critic = Critic(state_dim, action_dim).to(device)
        self.critic_target = copy.deepcopy(self.critic)
        self.critic_optimizer = torch.optim.Adam(self.critic.parameters(), lr=3e-4)

        # Hyperparams
        self.lr_decay = lr_decay
        self.grad_norm = grad_norm
        self.state_dim = state_dim
        self.max_action = max_action
        self.action_dim = action_dim
        self.discount = discount
        self.tau = tau
        self.eta = eta
        self.device = device
        self.max_q_backup = max_q_backup

    def step_ema(self):
        if self.step >= self.step_start_ema:
            self.ema.update_model_average(self.ema_model, self.actor)

    def train(self, replay_buffer, iterations, batch_size=64, log_writer=None):
        metric = {'bc_loss': [], 'ql_loss': [], 'actor_loss': [], 'critic_loss': []}
        for _ in range(iterations):
            # ===== Sample (supports PER and non-PER) =====
            sampled = replay_buffer.sample(batch_size)
            if len(sampled) == 5:
                state, action, next_state, reward, not_done = sampled
                weights = None
                idxs = None
            else:
                state, action, next_state, reward, not_done, weights, idxs = sampled  # PER path

            # >>>>>>> ADD THIS BLOCK to move to the right device <<<<<<<
            state     = state.to(self.device, non_blocking=True)
            action    = action.to(self.device, non_blocking=True)
            next_state= next_state.to(self.device, non_blocking=True)
            reward    = reward.to(self.device, non_blocking=True)
            not_done  = not_done.to(self.device, non_blocking=True)
            if weights is not None:
                weights = weights.to(self.device, non_blocking=True)

            # ----- Critic update (PER + PPO-style weighting) -----
            current_q1, current_q2 = self.critic(state, action)

            with torch.no_grad():
                if self.max_q_backup:
                    next_state_rep = torch.repeat_interleave(next_state, repeats=10, dim=0)
                    next_action_rep = self.ema_model(next_state_rep)
                    tq1, tq2 = self.critic_target(next_state_rep, next_action_rep)
                    tq1 = tq1.view(state.shape[0], 10).max(dim=1, keepdim=True)[0]
                    tq2 = tq2.view(state.shape[0], 10).max(dim=1, keepdim=True)[0]
                    target_q_min = torch.min(tq1, tq2)
                else:
                    next_action = self.ema_model(next_state)
                    tq1, tq2 = self.critic_target(next_state, next_action)
                    target_q_min = torch.min(tq1, tq2)
                target_q = (reward + not_done * self.discount * target_q_min).detach()

            # PPO-style regulator
            ppo_eps  = 0.10
            log_std  = -1.0
            reg_coef = 1.0
            sigma2 = torch.exp(2 * torch.tensor(log_std, device=self.device))

            def _logp_gauss(mu, a):
                return -0.5 * (((a - mu) ** 2).sum(dim=1, keepdim=True) / sigma2
                              + a.shape[1] * (np.log(2 * np.pi) + 2 * log_std))

            with torch.no_grad():
                mu_new = self.actor(state)
                mu_old = self.ema_model(state)
                logp_new = _logp_gauss(mu_new, action)
                logp_old = _logp_gauss(mu_old, action)
                ratio    = torch.exp(logp_new - logp_old)
                clip_r   = torch.clamp(ratio, 1.0 - ppo_eps, 1.0 + ppo_eps)

            td1 = current_q1 - target_q
            td2 = current_q2 - target_q
            mse = td1.pow(2) + td2.pow(2)

            if weights is not None:
                mse = mse * weights

            critic_loss = ((1.0 - reg_coef) * mse.mean()) + (reg_coef * (mse * clip_r).mean())

            self.critic_optimizer.zero_grad()
            critic_loss.backward()
            if self.grad_norm > 0:
                nn.utils.clip_grad_norm_(self.critic.parameters(), max_norm=self.grad_norm, norm_type=2)
            self.critic_optimizer.step()

            if idxs is not None:
                with torch.no_grad():
                    td_abs = 0.5 * (td1.abs() + td2.abs())
                replay_buffer.update_priorities(idxs, td_abs)

            # ----- Actor update (unchanged) -----
            bc_loss = self.actor.loss(action, state)
            new_action = self.actor(state)
            q1_new, q2_new = self.critic(state, new_action)
            if np.random.uniform() > 0.5:
                q_loss = - q1_new.mean() / (q2_new.abs().mean().detach() + 1e-8)
            else:
                q_loss = - q2_new.mean() / (q1_new.abs().mean().detach() + 1e-8)
            actor_loss = bc_loss + self.eta * q_loss

            self.actor_optimizer.zero_grad()
            actor_loss.backward()
            if self.grad_norm > 0:
                nn.utils.clip_grad_norm_(self.actor.parameters(), max_norm=self.grad_norm, norm_type=2)
            self.actor_optimizer.step()

            # ----- Targets & EMA -----
            if self.step % self.update_ema_every == 0:
                self.step_ema()
            with torch.no_grad():
                for p, tp in zip(self.critic.parameters(), self.critic_target.parameters()):
                    tp.data.mul_(1 - self.tau).add_(p.data, alpha=self.tau)
            self.step += 1

            # log
            metric['actor_loss'].append(float(actor_loss.item()))
            metric['bc_loss'].append(float(bc_loss.item()))
            metric['ql_loss'].append(float(q_loss.item()))
            metric['critic_loss'].append(float(critic_loss.item()))

        return metric

    @torch.no_grad()
    def sample_action(self, state_np):
        # state_np: flat numpy array
        state = torch.as_tensor(state_np.reshape(1, -1), dtype=torch.float32, device=self.device)
        # draw multiple actions, score by target critic, softmax sample index
        state_rep = torch.repeat_interleave(state, repeats=50, dim=0)
        action_batch = self.actor.sample(state_rep)
        q_vals = self.critic_target.q_min(state_rep, action_batch).flatten()
        probs = torch.softmax(q_vals, dim=0)
        idx = torch.multinomial(probs, 1)
        return action_batch[idx].cpu().numpy().flatten()

    def save_model(self, dir_path, id=None):
        import os
        os.makedirs(dir_path, exist_ok=True)
        name = f"actor_{id}.pth" if id is not None else "actor.pth"
        torch.save(self.actor.state_dict(), f"{dir_path}/{name}")
        name_c = f"critic_{id}.pth" if id is not None else "critic.pth"
        torch.save(self.critic.state_dict(), f"{dir_path}/{name_c}")

    def load_model(self, dir_path, id=None):
        name = f"actor_{id}.pth" if id is not None else "actor.pth"
        self.actor.load_state_dict(torch.load(f"{dir_path}/{name}", map_location=self.device))
        self.ema_model = copy.deepcopy(self.actor)
        name_c = f"critic_{id}.pth" if id is not None else "critic.pth"
        if os.path.exists(f"{dir_path}/{name_c}"):
            self.critic.load_state_dict(torch.load(f"{dir_path}/{name_c}", map_location=self.device))
            self.critic_target = copy.deepcopy(self.critic)
```

```python
# ==== training monitor ====
import numpy as np
import os
import csv
from collections import defaultdict

def train_with_monitoring(agent, replay_buffer, total_iters=10000, batch_size=64, log_interval=100,
                          out_dir="drive/MyDrive/stock_project/logs", run_name="run1"):
    """
    Trains in chunks of 'log_interval' and:
      - aggregates *per-step* losses returned by agent.train(...)
      - prints interval averages
      - returns the full history for plotting
      - (optional) writes CSV & npz at the end
    """
    os.makedirs(out_dir, exist_ok=True)
    hist = defaultdict(list)  # will store stepwise entries: actor_loss, critic_loss, bc_loss, ql_loss

    # Run
    for step in range(0, total_iters, log_interval):
        metrics = agent.train(replay_buffer, iterations=log_interval, batch_size=batch_size)

        # extend history (stepwise)
        for k, v in metrics.items():
            hist[k].extend(v)

        # interval averages
        avg_actor  = float(np.mean(metrics['actor_loss'])) if len(metrics['actor_loss']) else np.nan
        avg_critic = float(np.mean(metrics['critic_loss'])) if len(metrics['critic_loss']) else np.nan
        avg_bc     = float(np.mean(metrics['bc_loss']))    if len(metrics['bc_loss'])    else np.nan
        avg_ql     = float(np.mean(metrics['ql_loss']))    if len(metrics['ql_loss'])    else np.nan

        current_step = step + log_interval
        print(f"[{current_step}/{total_iters}] "
              f"Actor Loss: {avg_actor:.4f} | "
              f"Critic Loss: {avg_critic:.4f} | "
              f"BC Loss: {avg_bc:.4f} | "
              f"QL Loss: {avg_ql:.4f}")

    # ----- save CSV + NPZ for reproducibility -----
    steps = np.arange(len(hist['actor_loss']))
    csv_path = os.path.join(out_dir, f"{run_name}_training_log.csv")
    with open(csv_path, "w", newline="") as f:
        w = csv.writer(f)
        w.writerow(["step","actor_loss","critic_loss","bc_loss","ql_loss"])
        for i in range(len(steps)):
            w.writerow([
                int(steps[i]),
                float(hist['actor_loss'][i]),
                float(hist['critic_loss'][i]),
                float(hist['bc_loss'][i]),
                float(hist['ql_loss'][i]),
            ])
    npz_path = os.path.join(out_dir, f"{run_name}_training_log.npz")
    np.savez_compressed(npz_path,
                        step=steps,
                        actor_loss=np.array(hist['actor_loss'], dtype=np.float32),
                        critic_loss=np.array(hist['critic_loss'], dtype=np.float32),
                        bc_loss=np.array(hist['bc_loss'], dtype=np.float32),
                        ql_loss=np.array(hist['ql_loss'], dtype=np.float32))

    print(f"[saved] CSV: {csv_path}")
    print(f"[saved] NPZ: {npz_path}")

    return {
        "history": hist,           # dict(list)
        "steps": steps,            # numpy array
        "csv_path": csv_path,
        "npz_path": npz_path,
        "out_dir": out_dir,
        "run_name": run_name,
    }


# ===== use it exactly like before =====
iterations = 1000
batch_size = 64
run_name = "run1_lstm_per_ppoCritic"  # put a meaningful tag
result = train_with_monitoring(agent, per_buffer, total_iters=iterations, batch_size=batch_size,
                               log_interval=100, out_dir="drive/MyDrive/stock_project/logs", run_name=run_name)

# keep a handle to history for plotting
metrics = result["history"]
steps = result["steps"]
```

```python
# ==== Plotting with rolling averages & export ====
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

def _rolling(x, win=100):
    if len(x) == 0: return np.array([])
    win = min(win, len(x))
    c = np.cumsum(np.insert(x, 0, 0.0))
    y = (c[win:] - c[:-win]) / float(win)
    # pad to original len
    head = np.full(win-1, y[0]) if len(y) else np.array([])
    return np.concatenate([head, y]) if len(y) else np.array(x)

def plot_losses_for_thesis(history, steps, out_dir, run_name, roll=100):
    os.makedirs(out_dir, exist_ok=True)
    pdf_path = os.path.join(out_dir, f"{run_name}_loss_curves.pdf")
    png_path = os.path.join(out_dir, f"{run_name}_loss_curves.png")

    keys = ["actor_loss","critic_loss","bc_loss","ql_loss"]
    titles = {
        "actor_loss":  "Actor Loss (BC + Î·Â·QL)",
        "critic_loss": "Critic Loss (TD error)",
        "bc_loss":     "BC Loss (MSE / diffusion)",
        "ql_loss":     "QL Loss (âˆ’Q normalization)",
    }

    with PdfPages(pdf_path) as pdf:
        for k in keys:
            y = np.array(history[k], dtype=np.float32)
            ys = _rolling(y, win=roll)

            plt.figure(figsize=(10, 5))
            plt.plot(steps, y, label="raw")
            plt.plot(steps, ys, label=f"rolling mean (win={roll})", linewidth=2)
            plt.title(titles[k])
            plt.xlabel("Training step")
            plt.ylabel(k)
            plt.grid(True, alpha=0.3)
            plt.legend()
            plt.tight_layout()
            pdf.savefig()
            if k == "actor_loss":
                # save one representative PNG (actor loss) for quick preview
                plt.savefig(png_path, dpi=150)
            plt.show()
            plt.close()

    print(f"[saved] Loss curves PDF: {pdf_path}")
    print(f"[saved] Actor loss PNG: {png_path}")
    return pdf_path, png_path

pdf_path, png_path = plot_losses_for_thesis(metrics, steps,
                                            out_dir=result["out_dir"],
                                            run_name=result["run_name"],
                                            roll=100)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import matplotlib.dates as mdates
from matplotlib.backends.backend_pdf import PdfPages
import torch

# ===== Helper Metrics =====
def calculate_cumulative_return(portfolio_values):
    if len(portfolio_values) == 0:
        return 0
    initial_value = portfolio_values.iloc[0]
    final_value = portfolio_values.iloc[-1]
    return (final_value - initial_value) / initial_value

def calculate_max_earning_rate(portfolio_values, initial_cash):
    if len(portfolio_values) == 0:
        return 0
    return (portfolio_values.max() - initial_cash) / initial_cash

def calculate_max_pullback(portfolio_values):
    if len(portfolio_values) == 0:
        return 0
    peak = portfolio_values.iloc[0]
    drawdowns = []
    for value in portfolio_values:
        peak = max(peak, value)
        drawdowns.append((peak - value) / peak)
    return max(drawdowns)

def calculate_appt(portfolio_values, initial_cash, actions):
    if len(portfolio_values) == 0 or len(actions) == 0:
        return 0
    trades = np.diff(actions) != 0
    trade_indices = np.where(trades)[0] + 1
    trade_profits = []
    start_value = initial_cash
    for idx in trade_indices:
        if idx < len(portfolio_values):
            trade_profits.append(portfolio_values.iloc[idx] - start_value)
            start_value = portfolio_values.iloc[idx]
    return np.mean(trade_profits) if trade_profits else 0

def calculate_sharpe_ratio(returns, risk_free_rate=0.0, periods_per_year=252):
    if len(returns) == 0:
        return 0
    mean_return = np.mean(returns)
    std_return = np.std(returns)
    if std_return == 0:
        return 0
    annualized_return = mean_return * periods_per_year
    annualized_volatility = std_return * np.sqrt(periods_per_year)
    return (annualized_return - risk_free_rate) / annualized_volatility

# ===== Evaluate Diffusion-QL Agent =====
def evaluate_diffusion_agent(agent, test_env, initial_cash=1_000_000, pdf_path=None):
    obs, _ = test_env.reset()
    done = False

    while not done:
        action = agent.sample_action(obs)
        obs, reward, done, _, _ = test_env.step(action)

    total_values = test_env.get_total_values()
    actions = test_env.get_actions()

    portfolio_values = pd.Series(total_values, index=test_env.df.index[:len(total_values)])
    returns = portfolio_values.pct_change().dropna()

    metrics = {
        "Cumulative Return %": calculate_cumulative_return(portfolio_values) * 100,
        "Max Earning Rate %": calculate_max_earning_rate(portfolio_values, initial_cash) * 100,
        "Maximum Pullback %": calculate_max_pullback(portfolio_values) * 100,
        "APPT": calculate_appt(portfolio_values, initial_cash, np.array(actions)),
        "Sharpe Ratio": calculate_sharpe_ratio(returns)
    }

    # Plot cumulative return
    if pdf_path:
        with PdfPages(pdf_path) as pdf:
            plt.figure(figsize=(14, 8))
            plt.plot(portfolio_values.index, (portfolio_values - portfolio_values.iloc[0]) / portfolio_values.iloc[0], label='Our Agent')
            font_path = 'times new roman bold.ttf'
            font_prop = FontProperties(fname=font_path, size=14)
            plt.xlabel('Date', fontproperties=font_prop)
            plt.ylabel('Cumulative Return', fontproperties=font_prop)
            plt.legend(prop=font_prop)
            plt.grid(True)
            plt.gca().xaxis.set_major_locator(mdates.MonthLocator(bymonth=[1, 5, 9]))
            plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
            pdf.savefig()
            plt.show()

    return portfolio_values, metrics
```
